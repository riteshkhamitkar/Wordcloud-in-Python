```python
from wordcloud import WordCloud , STOPWORDS
```

```python
import numpy as np
```

```python
import PIL.Image
```

```python
import matplotlib.pyplot as plt
```

```python
text = open('mad.txt' , 'r').read()
```

```python
python_mask =np.array(PIL.Image.open("githublogo.png"))
```

```python
wc = WordCloud(stopwords = STOPWORDS, mask = python_mask , background_color="black").generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```
